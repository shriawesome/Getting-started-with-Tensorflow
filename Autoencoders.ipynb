```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

## Reading the Alphabets data
[Dataset link](http://kaggle.com/sachinpatel21/az-handwritten-alphabets-in-csv-format)
<br> It's a simple gray scale image value of handwritten aphabets, where each row corresponds to a gray scale intensity value. First column represents the labels

```python
alphabets_data=pd.read_csv('datasets/A_Z Handwritten Data.csv',header=None)
alphabets_data.sample(10)
```

## Understanding the Data

```python
# Size of the data
alphabets_data.shape
```

```python
# For now we'll work only with 5% of the data
alphabets_data=alphabets_data.sample(frac=0.05).reset_index(drop=True)
alphabets_data.shape
```

```python
# let's check the unique labels
sorted(alphabets_data[0].unique())
```

```python
# Let's setup a look up table for these mappings
lookup={0:'A',1:'B',2:'C',3:'D',4:'E',5:'F',6:'G',7:'H',8:'I',
        9:'J',10:'K',11:'L',12:'M',13:'N',14:'O',15:'P',16:'Q',
       17:'R',18:'S',19:'T',20:'U',21:'V',22:'W',23:'X',24:'Y',25:'Z'}
```

```python
# Since, Autoencoders are an Unsupervised learning algo we'll only extract
# and use features from our dataset.
features=alphabets_data[alphabets_data.columns[1:]]

target=alphabets_data[0]

features.shape
```

```python
# Each feature contains a 28x28 gray scale data
# let's reshape these values to 28 x 28 data.
features=features.values.reshape(features.shape[0],28,28)
features.shape
```

```python
# label at index 10, it's a O.
target.iloc[10]
```

```python
# Define a helper function to visualise an image
def show_img(features,actual_label):
    plt.imshow(features,cmap='Greys')
    plt.title('Actual label '+str(lookup[int(actual_label)]))
    plt.show()
```

```python
show_img(features[10],target[10])
```

```python
show_img(features[1000],target[1000])
```

```python
# Scaling the feature so that every image is on the scale 0-1
features=features.astype(np.float32)/255
```

## Splitting the data

```python
# Splitting the dataset into training and testing, with 80%
# data in training and the rest in testing.
train_img,test_img,train_label,test_label=train_test_split(features,
                                                           target,
                                                           test_size=0.2)

train_img.shape,test_img.shape
```

## Architecture for Autoencoders

### 1.  Architecture for the Encoder

```python
# For our Autoencoder, the first phase i.e. encoder will have 
# a dense layer, with the last layer as the coding layer.
stacked_encoder=tf.keras.Sequential([
    layers.Flatten(input_shape=[28,28]),
    layers.Dense(64,activation='relu'),
    layers.Dense(32,activation='relu'),
    # This layer holds the latent features
    layers.Dense(16,activation='relu')
])

stacked_encoder.summary()
```

### 3. Decoder Architecture

```python
# Decoder is an exact mirror image of the decoder, and the 
# o/p should the same as the input layer.
stacked_decoder=tf.keras.Sequential([
    layers.Dense(32,activation='relu',input_shape=[16]),
    layers.Dense(64,activation='relu'),
    layers.Dense(28*28,activation='relu'),
    
    layers.Reshape([28,28])
])
stacked_decoder.summary()
```

```python
# Now we can stack together these stacked layers together to form a 
# complete Autoencoder architecture.
ae_model=tf.keras.Sequential([stacked_encoder,stacked_decoder])

ae_model.summary()
```

```python
# For saving the architecture and visualising the nested sequential model
# we can use the plot_model's 'expand_nested' parameter.
keras.utils.plot_model(ae_model,expand_nested=True,show_shapes=True)
```

## Compiling and Training

```python
# compiling the model
ae_model.compile(optimizer=tf.keras.optimizers.RMSprop(),
                 loss=tf.keras.losses.MeanSquaredError(),
                 metrics=['mse'])
```

```python
# Since it is an unsupervised algo we we'll use labels as the training images itself,
# and also it's not  classification/regression thus no validation set.
training_hist=ae_model.fit(train_img,
                           train_img,
                           epochs=20)
```

## Visualising the result

```python
# Helper fn to visualise the o/p
def reconstruct_img(model,images,n_imgs):
    random_int=np.random.randint(0,images.shape[0]-n_imgs)
    
    reconstruction=model.predict(images[random_int:random_int+n_imgs])
    
    fig=plt.figure(figsize=(n_imgs*3,3))
    ax=fig.axes
    
    for img_idx in range(n_imgs):
        plt.subplot(2,n_imgs,1+img_idx)
        plt.imshow(images[random_int+img_idx],cmap='Greys')
        
        plt.subplot(2,n_imgs,1+n_imgs+img_idx)
        plt.imshow(reconstruction[img_idx],cmap='Greys')
```

```python
reconstruct_img(ae_model,test_img,5)
```

```python
# Different random imgs
reconstruct_img(ae_model,test_img,5)
```

Thus, we can see that the model does a pretty good job in reconstructing the data.

## Convolutional NN 
We know that the CNN's works better than regular NN when dealing with the image data.

### 1. Encoder Architecture

```python
conv_encoder=tf.keras.Sequential([
    # 1 for the Grey scale image(3 for RGB)
    layers.Reshape([28,28,1],input_shape=[28,28]),
    
    layers.Conv2D(16,(3,3),padding='same',activation='relu'),
    layers.MaxPooling2D(pool_size=(2,2)),
    
    layers.Conv2D(32,(3,3),padding='same',activation='relu'),
    layers.MaxPooling2D(pool_size=(2,2)),

    # Coding layer for our encoder
    layers.Conv2D(64,(3,3),padding='same',activation='relu'),
    layers.MaxPooling2D(pool_size=(2,2))
])
```

### 2. Decoder Architecture

```python
conv_decoder=tf.keras.Sequential([
    # for every Conv2D we have inverse layer for decoder
    layers.Conv2DTranspose(32,(3,3),strides=2,padding='valid',
                          activation='relu',input_shape=[3,3,64]),
    
    layers.Conv2DTranspose(16,(3,3),strides=2,padding='same',
                          activation='relu'),
    
    layers.Conv2DTranspose(1,(3,3),strides=2,padding='same',
                          activation='sigmoid'),
    
    layers.Reshape([28,28])
    
])
```

```python
# Stack together both these layers
conv_ae_model=tf.keras.Sequential([conv_encoder,conv_decoder])

conv_ae_model.summary()
```

```python
# Configuring parameters for the model
conv_ae_model.compile(loss=keras.losses.BinaryCrossentropy(),
                      optimizer=keras.optimizers.SGD(lr=1.0),
                      metrics=['mse'])
```

```python
training_hist_conv=conv_ae_model.fit(train_img,train_img,epochs=10)
```

## Visualising the result

```python
reconstruct_img(conv_ae_model,test_img,5)
```

```python
reconstruct_img(conv_ae_model,test_img,5)
```

Hence, we can clearly compare both the networks and can safely say that CNN's performs way better than the Dense NN.
